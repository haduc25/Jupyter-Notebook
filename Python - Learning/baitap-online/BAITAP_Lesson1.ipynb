Bài tập 3.8 Xây dựng lớp SinhVien gồm có các thuộc tính: Mã sinh viên, tên
sinh viên, năm sinh, quê quán.

    • Xây dựng phương thức nhập/hiện thông tin cho một sinh viên.
    • Xây dựng lớp quản lý sinh viên với các phương thức:
+ Nhập, xuất thông tin cho n sinh viên
+ Xây dựng phương thức in ra thông tin các sinh viên có quê quán ở Hải
Dương
+ Xây dựng phương thức in ra màn hình sinh viên có năm sinh nhỏ nhất.

```python
class SinhVien:
    # constructor
    def __init__(self, msv, tensv, ns, qq):
        self.msv = msv
        self.tensv = tensv
        self.ns = ns
        self.qq = qq
    
    def xuat(self):
        print('-'*25)
        print(f"Mã sinh viên: {self.msv}")
        print(f"Tên sinh viên: {self.tensv}")
        print(f"Năm sinh: {self.ns}")
        print(f"Quê quán: {self.qq}")
    
    def nhap(self):
        self.msv = input("\nNhập mã sinh viên: ")
        self.tensv = input("Nhập tên sinh viên: ")
        self.ns = input("Nhập năm sinh: ")
        self.qq = input("Nhập quê quán: ")

class QuanLySinhVien:
    def __init__(self):
        self.dssv = []
        
    def nhapDssv(self, n):
        for i in range(n):
            sv = SinhVien("", "", "", "")
            sv.nhap()
            self.dssv.append(sv)
            
    def xuatDssv(self):
        for sv in self.dssv:
            sv.xuat()
            
    def qqSinhVien(self, qq):
        for sv in self.dssv:
            if sv.qq == qq:
                sv.xuat()
        
    def nsSinhVienNhoNhat(self):
        sv_min = self.dssv[0]
        for sv in self.dssv:
            if sv.ns > sv_min.ns:
                sv_min = sv
        sv_min.xuat()


##################################################################
qlsv = QuanLySinhVien()

n = int(input("Nhập số lượng sinh viên: "))
qlsv.nhapDssv(n)

print("\n====> Danh sách sinh viên <====")
qlsv.xuatDssv()

print("\n=> Sinh viên có quê quán ở \'Hải Dương\'")
qlsv.qqSinhVien("Hải Dương")

print("\n=> Sinh viên có năm sinh nhỏ nhất")
qlsv.nsSinhVienNhoNhat()
```

Bài tập 3.9. Xây dựng lớp NhanVien gồm có các thuộc tính: Mã nhân viên,
tên nhân viên, quê quán, năm bắt đầu làm việc.

    • Xây dựng phương thức nhập/hiện thông tin cho một nhân viên.
    • Xây dựng phương thức in ra thông tin các nhân viên có quê quán ở
    Thái Nguyên
    • c. Xây dựng phương thức in ra màn hình nhân viên có thâm niên làm
    việc là ít nhất

```python
class NhanVien:
    def __init__(self, mnv, tennv, qq, nambatdau):
        self.mnv = mnv
        self.tennv = tennv
        self.qq = qq
        self.nambatdau = nambatdau 
    
    def xuat(self):
        print('-'*25)
        print(f"Mã nhân viên: {self.mnv}")
        print(f"Tên nhân viên: {self.tennv}")
        print(f"Năm bắt đầu làm việc: {self.nambatdau}")
        print(f"Quê quán: {self.qq}")
    
    def nhap(self):
        self.mnv = input("\nNhập mã nhân viên: ")
        self.tennv = input("Nhập tên nhân viên: ")
        self.qq = input("Nhập quê quán: ")
        self.nambatdau = input("Nhập năm bắt đầu làm việc: ")

        
class QuanLyNhanVien:
    def __init__(self):
        self.dsnv = []

    def nhapDsnv(self, n):
        for i in range(n):
            nv = NhanVien("", "", "", 0)
            nv.nhap()
            self.dsnv.append(nv)
            
    def xuatDsnv(self):
        for nv in self.dsnv:
            nv.xuat()

    def qqNhanVien(self, qq):
        for nv in self.dsnv:
            if nv.qq == qq:
                nv.xuat()

    def nhanVienLamViecItNhat(self):
        dsnv_itnhat = []
        nv_max = self.dsnv[0]
        for nv in self.dsnv:
            if nv.nambatdau > nv_max.nambatdau:
                nv_max = nv
        for nv in self.dsnv:
            if nv.nambatdau == nv_max.nambatdau:
                dsnv_itnhat.append(nv)
        for nv in dsnv_itnhat:
            nv.xuat()

    

##################################################################
qlnv = QuanLyNhanVien()

n = int(input("Nhập số lượng nhân viên: "))
qlnv.nhapDsnv(n)

print("\n====> Danh sách nhân viên <====")
qlnv.xuatDsnv()

print("\n=> Nhân viên có quê quán ở \'Thái Nguyên\'")
qlnv.qqNhanVien("Thái Nguyên")

print("\n=> Nhân viên có thâm niên làm việc là ít nhất")
qlnv.nhanVienLamViecItNhat()
```

Bài tập 3.10. Xây dựng lớp NhanVien gồm có các thuộc tính: Mã nhân viên,
tên nhân viên, năm sinh.

    • Xây dựng phương thức nhập/hiện thông tin cho một nhân viên.
    • Xây dựng lớp quản lý danh sách nhân viên với các phương thức:
        + Nhập và hiển thị danh sách n nhân viên
        + Phương thức lấy tuổi của nhân viên vừa nhập. (năm hiện tại – năm sinh)
        + Xây dựng phương thức in ra màn hình nhân viên có năm sinh cao nhất

```python
class NhanVien2:
    def __init__(self, mnv, tennv, ns):
        self.mnv = mnv
        self.tennv = tennv
        self.ns = ns
    
    def xuat(self):
        print('-'*25)
        print(f"Mã nhân viên: {self.mnv}")
        print(f"Tên nhân viên: {self.tennv}")
        print(f"Năm sinh: {self.ns}")
    
    def nhap(self):
        self.mnv = input("\nNhập mã nhân viên: ")
        self.tennv = input("Nhập tên nhân viên: ")
        self.ns = input("Nhập năm sinh: ")
    
    def getTuoi(self):
        return 2023 - int(self.ns)

        
class QuanLyNhanVien2:
    def __init__(self):
        self.dsnv = []

    def nhapDsnv(self, n):
        for i in range(n):
            nv = NhanVien2("", "", 0)
            nv.nhap()
            self.dsnv.append(nv)
            
    def xuatDsnv(self):
        for nv in self.dsnv:
            nv.xuat()
            print('Tuổi: ', nv.getTuoi())
        
    def nvTuoiCaoNhat(self):
        nv_max = self.dsnv[0]
        for nv in self.dsnv:
            if nv.getTuoi() > nv_max.getTuoi():
                nv_max = nv
        nv_max.xuat()
    

##################################################################
qlnv = QuanLyNhanVien2()

n = int(input("Nhập số lượng nhân viên: "))
qlnv.nhapDsnv(n)

print("\n====> Danh sách nhân viên <====")
qlnv.xuatDsnv()

print("\n=> Nhân viên có số tuổi cao nhất")
qlnv.nvTuoiCaoNhat()
```

Bài tập 3.11. Xây dựng lớp Sách với các thuộc tính: Tên sách, tác giả, giá
sách, năm xuất bản.

    • Xây dựng phương thức nhập/hiện thông tin cho lớp
    • Xây dựng lớp quản lý danh mục sách với các phương thức:
        + Nhập và hiển thị một danh sách
        + Xây dựng phương thức in ra màn hình thông tin về các sách có năm xuất
        bản lớn hơn 2005
        + Xây dựng phương thức in ra màn hình thông tin về sách có giá cao nhất

```python
class Sach:
    def __init__(self, tensach, tacgia, giasach, namxb):
        self.tensach = tensach
        self.tacgia = tacgia
        self.giasach = giasach
        self.namxb = namxb
    
    def xuat(self):
        print('-'*25)
        print(f"Tên sách: {self.tensach}")
        print(f"Tác giả: {self.tacgia}")
        print(f"Giá sách: {self.giasach}")
        print(f"Năm xuất bản: {self.namxb}")
    
    def nhap(self):
        self.tensach = input("\nNhập tên sách: ")
        self.tacgia = input("Nhập tên tác giả: ")
        self.giasach = int(input("Nhập giá sách: "))
        self.namxb = int(input("Nhập năm xuất bản: "))

class QuanLySach:
    def __init__(self):
        self.dms = []
        
    def nhapDms(self, n):
        for i in range(n):
            sach = Sach("", "", 0, 0)
            sach.nhap()
            self.dms.append(sach)
            
    def xuatDms(self):
        for sv in self.dms:
            sv.xuat()
            
    def sachCoNamXuatBanLonHonN(self, nam):
        print(f"\n=> Danh sách các sách xuất bản sau năm {nam}")
        for sach in self.dms:
            if sach.namxb > nam:
                sach.xuat()
                
    def sachGiaCaoNhat(self):
        sach_max = self.dms[0]
        for sach in self.dms:
            if sach.giasach > sach_max.giasach:
                sach_max = sach
        print("\n=> Sách có giá cao nhất là")
        sach_max.xuat()



##################################################################
qls = QuanLySach()

n = int(input("Nhập số lượng sách: "))
qls.nhapDms(n)

print("\n====> Danh mục sách <====")
qls.xuatDms()

qls.sachCoNamXuatBanLonHonN(2005)

qls.sachGiaCaoNhat()
```

Bài tập 3.12. Xây dựng lớp HocSinh gồm có các thuộc tính: ID, tên học sinh,
quê quán, điểm trung bình.

    • Xây dựng phương thức nhập/hiện thông tin cho một học sinh.
    • Xây dựng lớp quản lý sinh viên với các phương thức:
        + Nhập, xuất thông tin cho n học sinh
        + Xây dựng phương thức sắp xếp các học sinh theo thứ tự abc
        + Xây dựng phương thức in ra màn hình học sinh có điểm trung bình thấp
        nhất

```python
class HocSinh:
    def __init__(self, _id, _tenhs, _qq, _dtb):
        self.id = _id
        self.tenhs = _tenhs
        self.qq = _qq
        self.dtb = _dtb
    
    def xuat(self):
        print('-'*25)
        print(f"ID: {self.id}")
        print(f"Tên học sinh: {self.tenhs}")
        print(f"Quê quán: {self.qq}")
        print(f"Điểm trung bình: {self.dtb}")
    
    def nhap(self):
        self.id = input("\nNhập ID: ")
        self.tenhs = input("Nhập tên: ")
        self.qq = input("Nhập quê quán: ")
        self.dtb = float(input("Nhập điểm trung bình: "))
        
class QuanLyHocSinh:
    def __init__(self):
        self.dshs = []
        
    def nhapDshs(self, n):
        for i in range(n):
            hs = HocSinh("", "", "", 0)
            hs.nhap()
            self.dshs.append(hs)
            
    def xuatDshs(self):
        for hs in self.dshs:
            hs.xuat()
            
    def sapXepTheoTen(self):
        self.dshs.sort(key=lambda hs: hs.tenhs)
        
    def hsDiemThapNhat(self):
        hs_min = self.dshs[0]
        for hs in self.dshs:
            if hs.dtb < hs_min.dtb:
                hs_min = hs
        hs_min.xuat()



##################################################################
qlhs = QuanLyHocSinh()

n = int(input("Nhập số lượng học sinh: "))
qlhs.nhapDshs(n)

print("\n====> Danh sách học sinh <====")
qlhs.xuatDshs()

print("\n=> Danh sách học sinh sắp xếp theo tên")
qlhs.sapXepTheoTen()
qlhs.xuatDshs()

print("\n=> Hoc sinh có điểm trung bình thấp nhất")
qlhs.hsDiemThapNhat()
```
