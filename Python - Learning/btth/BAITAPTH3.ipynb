Bài tập 4.1. Viết chương trình cho phép:

    • Khởi tạo list
    • Thêm phần tử vào list
    • Nhập k, kiểm tra k xuất hiện bao nhiêu lần trong list
    • Tính tổng các số nguyên tố trong list
    • Sắp xếp
    • Xóa list

```python
import math

def khoiTaoList():
    myList = [2, 5, 7, 11]
    return myList

def thempt(myList, element):
    myList.append(element)

def demsopt(myList, element):
    count = myList.count(element)
    print(f"Số {element} xuất hiện trong list:", count, "lần")

def snt(num):
    if num > 1:
        for i in range(2, int(math.sqrt(num)) + 1):
            if num % i == 0:
                return False
        return True
    return False

def tongsnt(myList):
    prime_sum = 0
    for num in myList:
        if snt(num):
            prime_sum += num
    print("Tổng các số nguyên tố trong list là:", prime_sum)

def sxlist(myList):
    myList.sort()
    print("List sau khi sắp xếp:", myList)

def xoalist(myList):
    myList.clear()
    print("List sau khi xóa:", myList)

myList = khoiTaoList()
thempt(myList, 3)
thempt(myList, 13)
k = int(input('Nhập số k cần kiểm tra: '))
demsopt(myList, k)
tongsnt(myList)
sxlist(myList)
xoalist(myList)
```

Bài tập 4.2. Viết chương trình cho phép:
    
    • Viết lệnh khởi tạo ngẫu nhiên n phần tử cho list
    • Gọi k là một số nhập từ bàn phím, hãy xóa tất cả các phần tử có giá trị k tồn tại trong list
    • Kiểm tra list có đối xứng hay không

```python
import random

def khoiTaoList(n):
    myList = []
    for i in range(n):
        myList.append(random.randint(0, 10))
    return myList

def xoaPtTrongList(myList, k):
    myList = [elem for elem in myList if elem != k]
    return myList

def ktDoiXung(myList):
    n = len(myList)
    for i in range(n // 2):
        if myList[i] != myList[n - i - 1]:
            return False
    return True

n = 10
myList = khoiTaoList(n)
print("List ban đầu:", myList)

k = int(input("Nhập số k cần xóa: "))
myList = xoaPtTrongList(myList, k)
print("List sau khi xóa các phần tử có giá trị", k, "là:", myList)

if ktDoiXung(myList):
    print("List là đối xứng")
else:
    print("List không đối xứng")
```

Bài tập 4.3. 
Nhập từ bàn phím một số n nguyên dương, hãy tạo ra một danh
sách là list1 với n phần tử nhập từ bàn phím là số nguyên. 

Yêu cầu: Hãy tính
tổng số các phần tử chẵn và số các phần tử lẻ của list1

```python
def taoList(n):
    myList = []
    for i in range(n):
        elem = int(input("Nhập pt thứ [" + str(i+1) + "]: "))
        myList.append(elem)
    return myList

def demSoChanLe(myList):
    soLe = 0
    soChan = 0
    for elem in myList:
        if elem % 2 == 0:
            soChan += 1
        else:
            soLe += 1
    return (soLe, soChan)

n = int(input("Nhập số phần tử của list: "))
list1 = taoList(n)
print("List vừa nhập là:", list1)

soLe, soChan = demSoChanLe(list1)
print("Tổng số pt lẻ là:", soLe)
print("Tổng số pt chẵn là:", soChan)
```

Bài tập 4.4. Hãy viết hàm kiểm tra số nguyên tố. 

Yêu cầu: Nhập vào một số
nguyên dương n từ bàn phím, hãy liệt kê tất cả các số nguyên tố không vượt
quá n.

```python
def snt(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

n = int(input("Nhập n: "))

print(f"Các snt nhỏ hơn {n} là:")
for i in range(2, n+1):
    if snt(i):
        print(i)
```

Bài tập 4.5. Hãy nhập vào một số nguyên dương n từ bàn phím. 

In và liệt kê ra danh sách các ước số của n và đưa vào một danh sách có tên là list1.

```python
def timUocSo(n):
    uocso = []
    for i in range(1, n + 1):
        if n % i == 0:
            uocso.append(i)
    return uocso

n = int(input("Nhập số nguyên dương n: "))
uocso = timUocSo(n)
print(f"Các ước số của {n} là:", uocso)
list1 = uocso
print(f"Danh sách ước số của {n} là: {list1}")
```

Bài tập 4.7. Nhập lần lượt n số tự nhiên tạo thành một dãy a0, a1, ..., an. 

Gọi là dãy A. Hãy đưa ra một dãy con gọi là Bbao gồm các số nguyên tố của dãy A. 
In ra màn hình dãy B (Sử dụng sàng số nguyên tố Erastosthenes)

```python
import math

def eratosthenes(n):
    primes = [True] * (n+1)
    primes[0] = primes[1] = False
    for i in range(2, int(math.sqrt(n))+1):
        if primes[i]:
            for j in range(i**2, n+1, i):
                primes[j] = False
    return primes

n = int(input("Nhập số pt dãy A: "))
A = []
for i in range(n):
    num = int(input(f"Nhập pt A[{i+1}]: "))
    A.append(num)

is_prime = eratosthenes(max(A))
B = [num for num in A if is_prime[num]]

print("Dãy con B gồm các snt trong A:", B)
```

Bài tập 4.8. Cho trước một dãy A. Hãy in ra dãy B bằng cách tách tất cả các
phần tử là số của A. 

Ví dụ: A =[1.2, "one", 0.15, "aA", "B", 1.5], thì dãy B sẽ là: B = [1.2,0.15,1.5]

```python
def tachSo(A):
    B = []
    for x in A:
        if isinstance(x, (int, float)):
            B.append(x)
    return B

A = [1.2, "one", 0.15, "aA", "B", 1.5]
B = tachSo(A)
print(B)
```

Bài tập 4.9. Viết chương trình tính tổng các số trong một danh sách

Ví dụ:
Input: [2, 5, 8, 10, 12]
Output: Sum = 37

```python
def tongList(lst):
    sum = 0
    for num in lst:
        sum += num
    return sum

lst = [2, 5, 8, 10, 12]
sum = tongList(lst)
print("Sum =", sum)
```

Bài tập 4.10. Viết chương trình in ra tất cả các số chẵn trong một danh sách

Ví dụ:
Input: [2, 5, 8, 10, 12]
Output: 2, 8, 10, 12

```python
def soChan(numbers):
    for num in numbers:
        if num % 2 == 0:
            print(num, end=", ")

numbers = [2, 5, 8, 10, 12]
soChan(numbers)
```

Bài tập 4.11. Viết một chương trình tìm ra số lớn nhất của một danh sách
mà không sử dụng hàm max()

Ví dụ:
Input: [2, 5, 8, 10, 12]
Output: 12

```python
def timsolonnhat(lst):
    solonnhat = lst[0]
    for num in lst:
        if num > solonnhat:
            solonnhat = num
    return solonnhat

lst = [2, 5, 8, 10, 12]
solonnhat = timsolonnhat(lst)
print("Số max trong list:", solonnhat)
```

Bài tập 4.12. Viết một chương trình xóa tất cả phần tử lặp lại (trùng lặp) ra
khỏi danh sách.

Ví dụ:
Input: [1, 3, 5, 6, 3, 5, 6, 1]
Output: [1, 3, 5, 6]

```python
def xoaTrungLap(myList):
    newList = []
    for item in myList:
        if item not in newList:
            newList.append(item)
    return newList

lst = [1, 3, 5, 6, 3, 5, 6, 1]
print(lst)
newLst = xoaTrungLap(lst)
print(newLst)
```
