## *Building the Convolutional Neural Network (CNN) model*

```python
print('Building...')

# Building the CNN model
model = Sequential() #This line creates a sequential model object, which is a linear stack of layers.

model.add(Conv2D(32, (3, 3), padding='same', input_shape=(200,200,3), activation='relu'))
"""
=> model.add(Conv2D(32, (3, 3), padding='same', input_shape=(200,200,3), activation='relu'))
This line adds a convolutional layer to the model with 32 filters, each with a size of 3x3 pixels. 
The padding='same' argument ensures that the output feature maps have the same size as the input. 
The input_shape=(200,200,3) argument specifies the size and number of color channels of the input image. 
The activation='relu' argument specifies the rectified linear unit (ReLU) activation function.
"""
model.add(Conv2D(32, (3, 3), padding='same', activation='relu')) #This line adds another convolutional layer to the model with the same settings as the previous one.
model.add(MaxPool2D(pool_size=(2, 2)))
"""
=> model.add(MaxPool2D(pool_size=(2, 2)))
This line adds a max pooling layer to the model with a pool size of 2x2 pixels. 
This layer downsamples the output of the previous convolutional layer, reducing the spatial size of the feature maps.
"""

model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
"""
=> model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
=> model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
=> model.add(MaxPool2D(pool_size=(2, 2)))
These lines add two more sets of convolutional and max pooling layers to the model, but with 64 filters each.
"""


model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
"""
=> model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
=> model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
=> model.add(MaxPool2D(pool_size=(2, 2)))
These lines add two more sets of convolutional and max pooling layers to the model, but with 128 filters each.
"""

model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
"""
=> model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
=> model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
=> model.add(MaxPool2D(pool_size=(2, 2)))
These lines add two more sets of convolutional and max pooling layers to the model, but with 256 filters each.
"""

model.add(Flatten()) #This line adds a flatten layer to the model, which transforms the output of the previous layer into a 1D vector that can be fed into a dense layer.
model.add(Dense(units = 256, activation='relu'))
model.add(Dropout(0.5))
"""
=> model.add(Dense(units = 256, activation='relu'))
=> model.add(Dropout(0.5))
These lines add a dense layer to the model with 256 units and ReLU activation, followed by a dropout layer with a rate of 0.5. 
Dropout is a technique for preventing overfitting in neural networks by randomly dropping out (setting to zero) some of the input units.
"""

model.add(Dense(units = 256, activation='relu'))
model.add(Dropout(0.5))
"""
=> model.add(Dense(units = 256, activation='relu'))
=> model.add(Dropout(0.5))
These lines add another dense layer and dropout layer with the same settings as the previous ones.
"""

# configure the model before training
model.add(Dense(units = 1, activation = 'sigmoid')) #This line adds the final dense layer to the model with 1 unit and sigmoid activation, which is suitable for binary classification

model.compile(loss='binary_crossentropy',
            optimizer=RMSprop(learning_rate=0.0001),
            metrics=['accuracy'])
"""
=> model.compile(loss='binary_crossentropy',
            optimizer=RMSprop(learning_rate=0.0001),
            metrics=['accuracy'])
            
This code is setting up the Keras model for binary classification by specifying the loss function as binary_crossentropy, optimizer as RMSprop(lr=0.0001), and evaluation metric as accuracy. 
The compile() method is used to configure the model before training.
"""

print('done')
```

## *Performing the data augmentation to prevent overfitting*

```python
print('Preparing...')

# Preparing the data and performing Data Augmentation
train_datagen = ImageDataGenerator(rescale=1./255,
                                    shear_range=0.2,
                                    zoom_range=0.2,
                                    horizontal_flip=True)
"""
=> train_datagen = ImageDataGenerator(rescale=1./255,
                                    shear_range=0.2,
                                    zoom_range=0.2,
                                    horizontal_flip=True)
This line initializes an instance of the ImageDataGenerator class from the tensorflow.keras.preprocessing.image module,
which will be used to generate augmented data for the training set. 
The arguments rescale=1./255, shear_range=0.2, zoom_range=0.2, and horizontal_flip=True specify the types of augmentations to perform on the training images.

`rescale=1./255` scales the pixel values of the images to be between 0 and 1.
`shear_range=0.2` applies a random shear transformation with a maximum magnitude of 0.2 to each image.
`zoom_range=0.2` applies a random zoom transformation with a maximum magnitude of 0.2 to each image.
`horizontal_flip=True` randomly flips each image horizontally with a probability of 0.5.
"""

validation_datagen = ImageDataGenerator(rescale=1./255) #This line initializes an instance of the ImageDataGenerator class for the "validation set". The only augmentation applied to the validation images is rescale=1./255.

test_datagen = ImageDataGenerator(rescale=1./255) #This line initializes an instance of the ImageDataGenerator class for the "test set". Like the validation set, the only augmentation applied to the test images is rescale=1./255.

train_set = train_datagen.flow_from_directory('dataset/training_set',
                                                target_size=(200, 200),
                                                batch_size=64,
                                                class_mode='binary')
"""
=> train_set = train_datagen.flow_from_directory('dataset/training_set',
                                                target_size=(200, 200),
                                                batch_size=64,
                                                class_mode='binary')
                                                
This line generates batches of augmented training data from the directory "dataset/training_set". 
The flow_from_directory() method of the ImageDataGenerator class is used to create a generator that will load images from the specified directory, 
apply the augmentations specified by train_datagen, and return batches of images and their corresponding labels.

The arguments to flow_from_directory() are:
`target_size=(200, 200)`: resizes the images to be 200x200 pixels.
`batch_size=64`: returns batches of 64 images and labels.
`class_mode='binary'`: the labels are binary (i.e., 0 or 1).
"""

validation_set = validation_datagen.flow_from_directory('dataset/validation_set',
                                                target_size=(200, 200),
                                                batch_size=64,
                                                class_mode='binary')
"""
=> validation_set = validation_datagen.flow_from_directory('dataset/validation_set',
                                                target_size=(200, 200),
                                                batch_size=64,
                                                class_mode='binary')
                                                
This line generates batches of validation data using the same method as train_set, but for the directory "dataset/validation_set". 
The arguments are the same as for train_set.
"""
### START - Reduce change path ###
# test_set = test_datagen.flow_from_directory('dataset/testing_set',
#                                                 target_size=(200, 200),
#                                                 batch_size=1,class_mode = 'binary',shuffle = False)
"""
=> test_set = test_datagen.flow_from_directory('dataset/testing_set',
                                                target_size=(200, 200),
                                                batch_size=1,class_mode = 'binary',shuffle = False)
                                                
This line generates batches of test data using the same method as train_set, but for the directory "dataset/testing_set". 
The arguments are the same as for train_set, except for batch_size=1 and shuffle=False.

`batch_size=1` returns individual images rather than batches.
`shuffle=False` ensures that the images are returned in the same order as they appear in the directory
"""
### END - Reduce change path ###

### START - new code ###
test_set = test_datagen.flow_from_directory('kaggle/working/dataset/testing_set',
                                                target_size=(200, 200),
                                                batch_size=1,class_mode = 'binary',shuffle = False)
### END - new code ###



print('done')
```

## *Building the Callback Functions*

```python
print('Creating the callback functions...')

# Creating the callback functions
# EarlyStopping callback:
es = EarlyStopping(monitor='val_loss',mode= 'min',verbose=1,patience=10,restore_best_weights=True)
"""
=> es = EarlyStopping(monitor='val_loss',mode= 'min',verbose=1,patience=10,restore_best_weights=True)

This line creates an EarlyStopping callback object. 
This callback is used to stop training the model when the validation loss metric stops improving. 
=> Trnh ci tnh trng m b overfitting 

The parameters are:
`monitor:` This parameter specifies the metric to monitor for early stopping. In this case, it is set to 'val_loss', which means that the validation loss will be monitored. The training process will stop if the validation loss stops improving.
`mode:` This parameter specifies whether to minimize or maximize the monitored metric. In this case, it is set to 'min', which means that the validation loss should be minimized.
`verbose:` This parameter specifies how much output to produce during the training process. A value of 1 means that a message will be printed when early stopping is triggered.
`patience:` This parameter specifies how many epochs to wait before stopping training if the monitored metric does not improve. In this case, it is set to 10, which means that training will stop if the validation loss does not improve for 10 epochs.
`restore_best_weights:` This parameter specifies whether to restore the weights of the model to the best performing epoch when training is stopped due to early stopping. In this case, it is set to True, which means that the weights of the best performing epoch will be used.
"""
print('EarlyStopping callback is successfully!')

# ModelCheckpoint callback:
mc = ModelCheckpoint('best_model.h5',monitor='val_loss',mode= 'min',verbose=1,save_best_only=False)
"""
=> mc = ModelCheckpoint('best_model.h5',monitor='val_loss',mode= 'min',verbose=1,save_best_only=False)
This line creates a ModelCheckpoint callback object. 
This callback is used to save the weights of the model during training, so that the best performing epoch can be used later. 
=> Lu li `weights of the model` nu c b overfitting th c th quay li c

The parameters are:
`filepath:` This parameter specifies the name of the file to save the weights of the model to. In this case, it is set to 'best_model.h5'.
`monitor:` This parameter specifies the metric to monitor for improvement. In this case, it is set to 'val_loss', which means that the validation loss will be monitored.
`mode:` This parameter specifies whether to minimize or maximize the monitored metric. In this case, it is set to 'min', which means that the validation loss should be minimized.
`verbose:` This parameter specifies how much output to produce during the training process. A value of 1 means that a message will be printed when the model weights are saved.
`save_best_only:` This parameter specifies whether to save only the weights of the best performing epoch or all epochs. In this case, it is set to False, which means that the weights of all epochs will be saved.
"""
print('ModelCheckpoint callback is successfully!')

# CSVLogger callback:
csv = CSVLogger('CSVLogs.csv',separator=',', append=False)
"""
=> csv = CSVLogger('CSVLogs.csv',separator=',', append=False)

=> This line creates a CSVLogger callback object. This callback is used to save the training and validation metrics to a CSV file during training. 

The parameters are:

`filename:` This parameter specifies the name of the CSV file to save the training and validation metrics to. In this case, it is set to 'CSVLogs.csv'.
`separator:` This parameter specifies the character used to separate the values in the CSV file. In this case, it is set to ','.
`append:` This parameter specifies whether to append to an existing CSV file or overwrite it. In this case, it is set to False, which means that the file will be overwritten.
"""

print('CSVLogger callback is successfully!')
```

```python
print('Fitting...')
# Fitting the model to data
### START - change fit_generator method to fit method ###
# output = model.fit_generator(train_set, steps_per_epoch=(18697//64),epochs= 20,
#                              validation_data = validation_set, validation_steps= (6303//64),
#                              callbacks=[es,mc,csv,PlotLossesKeras()])
"""
=> output = model.fit_generator(train_set, steps_per_epoch=(18697//64),epochs= 20,
                             validation_data = validation_set, validation_steps= (6303//64),
                             callbacks=[es,mc,csv,PlotLossesKeras()])
                             
`output = model.fit_generator(train_set, ...)` - This line fits the model to the training data train_set using the fit_generator method. The fit_generator method is used to train models on large datasets that cannot be loaded into memory all at once.
`steps_per_epoch=(18697//64)` - This parameter is used to define the number of training steps per epoch. In this case, it is set to 18697//64, which is the number of samples in the training dataset divided by the batch size of 64. This parameter determines how many batches of samples are processed before one epoch is completed.
`epochs=20` - This parameter defines the number of epochs (i.e., complete passes through the entire training dataset) for which the model will be trained.
`validation_data = validation_set` - This parameter specifies the validation data that will be used during the training process. validation_set is typically a set of data that is used to evaluate the performance of the model during training, and it is usually different from the training data.
`validation_steps= (6303//64)` - This parameter is similar to steps_per_epoch, but it is used for the validation dataset. It is set to 6303//64, which is the number of samples in the validation dataset divided by the batch size of 64.
`callbacks=[es,mc,csv,PlotLossesKeras()]` - This parameter is a list of callback functions that are called during the training process. Callbacks are used to customize the behavior of the training process. In this case, four callbacks are used:
    `es` is an early stopping callback that stops the training process if the validation loss does not improve for a certain number of epochs.
    `mc` is a model checkpoint callback that saves the weights of the model at certain intervals during the training process.
    `csv` is a CSV logger callback that logs the training and validation metrics to a CSV file.
    `PlotLossesKeras()` is a custom callback that plots the training and validation loss during training.
=> es,mc,csv Cc callback  c khai bao  bn trn
"""
### END - change fit_generator method to fit method ###

### START - using fit method ###
output = model.fit(train_set, steps_per_epoch=(18697//64), epochs=20,
                   validation_data=validation_set, validation_steps=(6303//64),
                   callbacks=[es, mc, csv, PlotLossesKeras()])
### END - using fit method ###
print('done')
```

## *Loading the VGG16, VGG19 model for transfer learning* (Using VGG16)

```python
print('Loading the VGG model...')
# Loading the VGG model using Transfer Learning
transfer_model = VGG16(include_top = False, weights = 'imagenet',input_shape = (200,200,3))
for layer in transfer_model.layers:
    layer.trainable = False
    
# Defining the output layers
flat = Flatten()(transfer_model.layers[-1].output)
class1 = Dense(units = 256, activation='relu')(flat)
class2 = Dense(units = 256, activation = 'relu')(class1)
output = Dense(units = 1, activation='sigmoid')(class2)

# define new model
transfer_model = Model(inputs=transfer_model.inputs, outputs=output)
transfer_model.compile(loss='binary_crossentropy',
#             optimizer=RMSprop(lr=0.001),
            optimizer=RMSprop(learning_rate=0.001), # lr => learning_rate
            metrics=['accuracy'])

"""
=> transfer_model = VGG16(include_top = False, weights = 'imagenet',input_shape = (200,200,3))

This line loads the VGG16 pre-trained model from Keras with include_top = False to exclude the fully-connected layer at the top,
weights = 'imagenet' to use pre-trained weights, 
and input_shape = (200,200,3) to set the input shape of the model to (200, 200, 3) for RGB images.

=> for layer in transfer_model.layers:
    layer.trainable = False

This code sets all layers in the pre-trained model as non-trainable so that their weights remain fixed.


=>  flat = Flatten()(transfer_model.layers[-1].output)
    class1 = Dense(units = 256, activation='relu')(flat)
    class2 = Dense(units = 256, activation = 'relu')(class1)
    output = Dense(units = 1, activation='sigmoid')(class2)

These lines define the output layers of the new model.

`Flatten()` layer takes the output of the last layer in the pre-trained model as input and flattens it into a 1D vector. | chuyn v vector
`Dense()` layer with 256 units and ReLU activation function is added on top of the flattened layer `flat` to perform feature extraction.
Another `Dense()` layer with 256 units and ReLU activation function is added to further learn high-level features.
The final output layer is a `Dense()` layer with 1 unit and sigmoid activation function, which outputs a probability score between 0 and 1 representing the predicted class probability.

=>  transfer_model = Model(inputs=transfer_model.inputs, outputs=output)
    transfer_model.compile(loss='binary_crossentropy',
                optimizer=RMSprop(lr=0.001),
                metrics=['accuracy'])

These lines define the new model by setting the input and output layers. 
The `Model()` method is used to define the new model using the pre-trained model's inputs and the newly defined output layer `output`.

The `compile()` method is then used to configure the learning process of the new model. 
The `loss` parameter is set to `'binary_crossentropy'` since this is a binary classification problem. 
The `optimizer` parameter is set to use the `RMSprop` optimizer with a learning rate of 0.001. 
Finally, the `metrics` parameter is set to `'accuracy'` to monitor the of the model during training.
"""

print('done')
```
